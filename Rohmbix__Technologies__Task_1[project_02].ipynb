```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data: pd.DataFrame = pd.read_csv("cleaned_loan_prediction_data.csv")
```

```python
                               # Check Correlation
data.corr(numeric_only=True)
```

```python
plt.figure()          # heatmap shows correlation between variables
sns.heatmap(data.corr(numeric_only=True), annot=True)
plt.show()
```

```python
                                # Detect Outliers
plt.figure()
sns.boxplot(x=data['LoanAmount'])
plt.show()
```

```python
                              # Create Visualizations
```

```python
        # 1- Histogram
plt.figure()
data['LoanAmount'].hist()
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
       # 2- BarChart
plt.figure()
data['CreditScore'].value_counts().head(10).plot(kind='bar')
plt.title("Top Credit Scores")
plt.show()
```

```python
        # 3- Line Graph
plt.figure()
data['LoanAmount'].head(20).plot()
plt.title("Loan Amount Trend")
plt.show()
```
